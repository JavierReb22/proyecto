```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x)else x)
```

```python
df_DA_ch= df[(df['job_title_short']== 'Data Analyst') & (df['job_country'] == 'Chile')].copy()
```

```python
df_DA_ch['salary_year_avg'].plot(kind='hist',bins=30, edgecolor='black')
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x , pos: f'${int(x/1000)}K'))
plt.xlim(0,250000)
plt.title('Distribucion de salarios de Data Analyst en Chile')
plt.xlabel('Salario Anual')
plt.ylabel('Cantidad de trabajos')
plt.show()
```
